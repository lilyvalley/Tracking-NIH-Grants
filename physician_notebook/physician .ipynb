```python
d2.head()
```

```python
d4.head(10)
```

```python
d5 = d4[d4["Measure Performance Rate"]>10]
```

```python
print(d4.shape)
print(d5.shape)
```

```python
d5.head()
```

```python
d6 = d5.groupby("Graduation year").mean()
```

```python
d6.head(10)
d7 = d6.reset_index()
d7.head()
```

```python
## linear regression

x, y = d7["Graduation year"],d7["Measure Performance Rate"]
```

```python
from scipy import stats
slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)
```

```python
print(slope,p_value)
```
